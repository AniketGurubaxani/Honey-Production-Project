#### 3. Display the percentage distribution of the data in each year using the pie chart.

```python
plt.pie(df["year"].value_counts(),labels = df["year"].value_counts().index, autopct='%1.1f%%')
plt.title('Percentage Distribution of Data in Each Year');
```

#### 4. Plot and Understand the distribution of the variable "price per lb" using displot, and write your findings.

```python
sns.displot(df["priceperlb"], kde = True, bins = 15)
plt.xlabel('Price per lb');
```

### <font color = blue> *Inferences -* </font>
- Above distribution plot shows Price per lb is positively skewed variable.
- It can be seen that most of average price per pound lies in range of 0.5 dollars to 1.5 dollars.

#### 5. Plot and understand the relationship between the variables 'numcol' and 'prodval' through scatterplot, and write your findings.

```python
plt.title("Relationship between Number of honey producing colonies & Value of production.");
sns.scatterplot(df, x = 'numcol', y = 'prodvalue');
plt.xlabel("Number of honey producing colonies")
plt.ylabel("Value of production");
```

### <font color = blue> *Inferences -* </font>
- Above scatter plot shows the good positive correlation between "Number of honey producing colonies" and "Value of production".
- They are directly proportional to each other.
- As No. of colonies producing honey increases, production also get increases.

#### 6. Plot and understand the relationship between categorical variable 'year' and a numerical variable 'prodvalue' through boxplot, and write your findings.

```python
plt.figure(figsize=(10,5))
plt.title("Value of production in each year")
sns.boxplot(df, x = "year", y = "prodvalue");
```

### <font color = blue> *Inference -* </font>
- From above box plot it can be noticed that maximum production is done in year 2002, 2011, 2012 which are led by the outliers.
- Production value having only outliers which are above the upper bound.

#### 7. Visualize and understand the relationship between the multiple pairs of variables throughout different years using pairplot and add your inferences. (use columns 'numcol', 'yield percol', 'total prod', 'prodvalue','year')

```python
attr = ['numcol', 'yieldpercol', 'totalprod', 'prodvalue', "year"]
```

```python
sns.pairplot(df[attr], hue = "year", palette = "Spectral");
plt.suptitle("Relationship between the multiple pairs of variables throughout different years", y =1 );
```

### <font color = blue> *Inference -* </font>
- Production value and no. of colonies are high from year 2000 to 2005 and also sometimes high in years 2010 and 2012.
- Most of the production value lies in 0 to 2 dollars.
- Honey produced between range of 25 lbs to 75 lbs in year 2012.
- Production of honey is maximum in previous years like 2000 and 2002. Amount produced in these years is between range of 100 lbs to 150.

#### 8. Display the correlation values using a plot and add your inferences. (use columns 'numcol', 'yield percol','total prod', 'stocks', 'price per lb', 'prodvalue').
